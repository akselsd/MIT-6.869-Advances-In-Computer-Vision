```python
def imshow1(image, show_axes = False, quiet = False):
    if len(image.shape) == 3:
      # Height, width, channels
      # Assume BGR, do a conversion since 
      image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:
      # Height, width - must be grayscale
      # convert to RGB, since matplotlib will plot in a weird colormap (instead of black = 0, white = 1)
      image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    # Draw the image
    plt.imshow(image)
    if not show_axes:
        # We'll also disable drawing the axes and tick marks in the plot, since it's actually an image
        plt.axis('off')
    if not quiet:
        # Make sure it outputs
        plt.show()
```

```python
kim_dir = os.path.join(rootpath, "images", "kim")
trump_dir = os.path.join(rootpath, "images", "trump")

def fn_to_im_gs(img_dir, fn):
    return cv2.imread(os.path.join(img_dir, fn), cv2.IMREAD_GRAYSCALE)
    

def gaussian_blur_kernel(sd):
    # Use  [-3sd, 3sd] interval size
    size = 2*3*sd + 1
    size = int(size)
    if size % 2 == 0:
        size = size + 1
    kernel = np.zeros((size, size))
    offset = int((size - 1)/2)
    for m in range(size):
        for n in range(size):
            x = m - offset
            y = n - offset
            kernel[m][n] = np.exp(-(x*x + y*y)/(2*sd*sd))
    return kernel / np.sum(kernel)
    
def box_blur_kernel(size):
    if size % 2 == 0:
        size = size + 1
    return np.ones((size, size))/(size**2)


def hybrid_img(imgA_raw, imgB_raw, kernel):      
    size = min(imgA_raw.shape[0], imgB_raw.shape[0], 400)
    
    imgA = cv2.resize(imgA_raw, (size, size)).astype(np.float32)
    imgB = cv2.resize(imgB_raw, (size, size)).astype(np.float32)
    
    out = np.zeros((imgA.shape))
    
    edge_strat = "fill" #fill, wrap, symm
    out = 0.5*scipy.signal.convolve2d(imgB, kernel, boundary=edge_strat, mode="same")
    out = out + 0.5*(imgA - scipy.signal.convolve2d(imgA, kernel, boundary=edge_strat, mode="same"))
    
    #Unsure if this will be a problem, but just truncate in case it is
    out[out < 0] = 0
    out[out > 255] = 255

    return out.astype(np.uint8)

N_KIMS = 5
N_TRUMPS = 8

kim_fns = ["kim" + str(i + 1) + ".jpg" for i in range(N_KIMS)]
trump_fns = ["trump" + str(i + 1) + ".jpg" for i in range(N_TRUMPS)]

kim_imgs = [fn_to_im_gs(kim_dir, fn) for fn in kim_fns]
trump_imgs = [fn_to_im_gs(trump_dir, fn) for fn in trump_fns]

"""
Some good ones: 
Box(9), Trump 6, Kim 0
Box(9), Trump 0, Kim 2
Gaussian(2.3) Trump 2, Kim 2

"""

kernel = box_blur_kernel(9)
print("Box kernel of size 9")
imshow1(hybrid_img(kim_imgs[0], trump_imgs[6], kernel))
print("Box kernel of size 9")
imshow1(hybrid_img(kim_imgs[2], trump_imgs[0], kernel))
print("Gaussian kernel with sd = 2.3")
kernel = gaussian_blur_kernel(2.3)
imshow1(hybrid_img(kim_imgs[2], trump_imgs[2], kernel))


for A in kim_imgs:
    for B in trump_imgs:
        pass
        #imshow1(hybrid_img(A, B, kernel)

kernel = gaussian_blur_kernel(10)        
print("High blur (gaussian kernel with sd 10)")
imshow1(hybrid_img(kim_imgs[2], trump_imgs[2], kernel))
kernel = gaussian_blur_kernel(1)   
print("Low blur (gaussian kernel with sd 1)")
imshow1(hybrid_img(kim_imgs[2], trump_imgs[2], kernel))
```

```python
### scale image's mean intensity to 127 for better visualization.
def intensityscale(raw_img):
    
    # scale an image's intensity from [min, max] to [0, 255].
    v_min, v_max = raw_img.min(), raw_img.max()
    scaled_im = (raw_img * 1.0 - v_min) / (v_max - v_min) * 255.
    
    # keep the mean to be 127.
    meangray = np.mean(scaled_im)
    scaled_im = scaled_im - meangray + 127.
    
    # clip to [0, 255]
    scaled_im = np.clip(scaled_im, 0, 255)
    
    return scaled_im.astype(np.uint8)

raw_img = cv2.imread(os.path.join(rootpath, "einsteinandwho.jpg"), cv2.IMREAD_GRAYSCALE)
f_img = np.fft.fftshift(np.fft.fft2(raw_img))

M, N = raw_img.shape

xfreq = np.fft.fftfreq(f_img.shape[0])
yfreq = np.fft.fftfreq(f_img.shape[1])

def circle_matrix(threshold):
    dists = np.arange(-int(M/2), int(M/2))**2
    sq_dists = np.sqrt(dists[:, np.newaxis] + dists)
    return (np.abs(sq_dists) < (M/2)*threshold).astype(int)


def exp_matrix(gamma):
    dists = np.arange(-int(M/2), int(M/2))**2
    sq_dists = np.sqrt(dists[:, np.newaxis] + dists)
    return np.exp(-gamma*np.abs(sq_dists))



    
filters = [exp_matrix(50/M),
           circle_matrix(0.06),
           0.5*exp_matrix(50/M) + 0.5*circle_matrix(0.06),
           np.multiply(exp_matrix(50/M), circle_matrix(0.08))]
    
for img_filter in filters:

    low = img_filter
    high = np.abs(1 - low)

    f_img_low = np.fft.ifftshift(f_img*low)
    f_img_high = np.fft.ifftshift(f_img*high)

    high_img = np.real(np.fft.ifft2(f_img_high))
    low_img = np.real(np.fft.ifft2(f_img_low))
    
    high_img = intensityscale(high_img)
    low_img = intensityscale(low_img)
    
    imshow1(np.hstack((low_img, high_img)))
```

### Magnify Change
Fill out code **here**

```python
def magnifyChange(im1, im2, magnificationFactor):
    
    # find phase shift in frequency domain
    im1Dft = fft2(im1)
    im2Dft = fft2(im2)
    phaseShift = np.angle(im2Dft) - np.angle(im1Dft)
    
    # magnify the phase change in frequency domain
    magnifiedDft = np.abs(im2Dft)*np.exp(1j*(phaseShift*magnificationFactor + np.angle(im1Dft)))
    
    # what does the magnified phase change cause in image space?
    magnified = ifft2(magnifiedDft).real;
    
    return magnified
```

### Results

If we want this to work, we need that the movement in one direction of one pixel does not affect the magnified movement of the other. However, since the total phase shift is not the sum of the phase shifts caused by moving the two pixels (taking the angle is not a linear operator)
$$ \angle \mathcal{F}(im1 + im2) \ne \angle \mathcal{F}(im1) + \angle \mathcal{F}(im2) $$

we get this effect that both pixels have their movements magnified both up and to the right (wrapping around), regardless of that pixels initial movement. That is why the window idea works, we limit the phase shift calculation to only consider the local area in the window. 

## Problem 3.c
Fill out code **here**

```python
# 9x9 images
imSize = 9

# we would like to magnify the change between im1 and im2 by 4x
magnificationFactor = 4

# width of our Gaussian window
sigma = 2

# horizontal movement from (1, 1) to (1, 2)
# additional vertical movement from (9, 9) to (8, 9)
im1 = np.zeros([imSize, imSize])
im2 = np.zeros([imSize, imSize])
im1[0,0] = 1
im2[0,1] = 1
im1[8,8] = 1
im2[7,8] = 1

# we will magnify windows of the image and aggregate the results
magnified = np.zeros([imSize, imSize])

# meshgrid for computing Gaussian window
X, Y = np.meshgrid(np.arange(imSize), np.arange(imSize))

for y in range(0, imSize, 2*sigma):
    for x in range(0, imSize, 2*sigma):
        gaussianMask = np.exp(-0.5 * (X - x)**2 / sigma**2)*np.exp(-0.5 * (Y - y)**2  / sigma**2)
        windowMagnified = magnifyChange(np.multiply(im1, gaussianMask), np.multiply(im2, gaussianMask)
                        ,magnificationFactor)
        magnified = magnified + windowMagnified
        
plt.figure(figsize=(12,36))
plt.subplot(131)
imshow(im1); plt.title('im1');

plt.subplot(132)
imshow(im2); plt.title('im2');

plt.subplot(133)
imshow(magnified); plt.title('magnified');
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture('bill.avi')

# list of video frames
frames = []

while(cap.isOpened()):
    # read frame from the video
    ret, frame = cap.read()
    
    if ret is False:
        break
        
    frames.append(frame)

cap.release()

# scale frame to 0-1
frames = np.array(frames) / 255.
print("frames size:", frames.shape, "# (nb_frames, height, width, channel)")

# get height, width
numFrames = frames.shape[0]
height = frames.shape[1]
width = frames.shape[2]
```

### Motion magnification
Fill out code **here**

```python
# 10x magnification of motion
magnificationFactor = 10

# width of Gaussian window
sigma = 13

# alpha for moving average
alpha = 0.5

# we will magnify windows of the video and aggregate the results
magnified = np.zeros_like(frames)

# meshgrid for computing Gaussian window
X, Y = np.meshgrid(np.arange(width), np.arange(height))

# iterate over windows of the frames
xRange = list(range(0, width, 2*sigma))
yRange = list(range(0, height, 2*sigma))
numWindows = len(xRange) * len(yRange)
windowIndex = 1

for y in yRange:
    for x in xRange:
        for channelIndex in range(3): # RGB channels
            for frameIndex in range(numFrames):
                
                # create windowed frames
                gaussianMask = np.exp( ((X - x)**2 + (Y - y)**2)  / (-2*sigma**2))
                windowedFrames = gaussianMask * frames[frameIndex,:,:,channelIndex]
            
                # initialize moving average of phase for current window/channel
                if frameIndex == 0:
                    windowAveragePhase = angle(fft2(windowedFrames))
                
                windowDft = fft2(windowedFrames)
                
                # compute phase shift and constrain to [-pi, pi] since
                # angle space wraps around
                windowPhaseShift = angle(windowDft) - windowAveragePhase
                windowPhaseShift[windowPhaseShift > pi] = windowPhaseShift[windowPhaseShift > pi] - 2 * pi
                windowPhaseShift[windowPhaseShift < -pi] = windowPhaseShift[windowPhaseShift < -pi] + 2 * pi
                
                # magnify phase shift
                windowMagnifiedPhase = windowPhaseShift*magnificationFactor + windowAveragePhase
                 
                # go back to image space
                windowMagnifiedDft = np.abs(windowDft)*np.exp(1j*(windowMagnifiedPhase))
                windowMagnified = abs(ifft2(windowMagnifiedDft))
                
                # update moving average
                windowPhaseUnwrapped = windowAveragePhase + windowPhaseShift
                windowAveragePhase = alpha * windowAveragePhase + (1 - alpha) * windowPhaseUnwrapped
                
                # aggregate
                magnified[frameIndex,:,:,channelIndex] = magnified[frameIndex,:,:,channelIndex] + windowMagnified
        
        # print progress
        print('{}/{} x:{} y:{} '.format(windowIndex, numWindows, x, y), end='\r')
        windowIndex += 1
```

```python
outputs = magnified / np.max(magnified)
for channelIndex in range(3):
    originalFrame = frames[0,:,:,channelIndex]
    magnifiedFrame = outputs[0,:,:,channelIndex]
    scale = np.std(originalFrame[:]) / np.std(magnifiedFrame[:])
    originalMean = np.mean(originalFrame[:])
    magnifiedMean = np.mean(magnifiedFrame[:])
    outputs[:,:,:,channelIndex] = magnifiedMean + scale * (outputs[:,:,:,channelIndex] - magnifiedMean)

outputs = np.clip(outputs, 0, 1)
```

```python
# create output video
fourcc = cv2.VideoWriter_fourcc('M','J','P','G')
# fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('bill_magnified.avi',fourcc, 30.0, (height, width))

for i in range(frames.shape[0]):
    # scale the frame back to 0-255
    frame = (np.clip(outputs[i], 0, 1) * 255).astype(np.uint8)
    
    # write frame to output video
    out.write(frame)

out.release()
```
